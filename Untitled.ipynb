```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import torch
from torch import nn, optim
from torchvision import models, transforms, utils
from PIL import Image
```

```python
IMGPATH = 'images_training_rev1'
imglist = [f'{IMGPATH}/{i}' for i in os.listdir(IMGPATH)]
```

```python
def get_image(n=6):
    """ Get n random images, each an |m x n x 3| array, from training set
    """
    fnames = np.random.choice(imglist, size=n, replace=False)
    return np.array([np.array(Image.open(f)) for f in fnames])
    
```

```python
def figpos(n_cols, i):
    """ Given a number, return its corresponding coordinate in a 2D grid
    """
    row = i // n_cols
    col = i % n_cols
    
    return row, col

def show(arr, n_cols=3):
    """ Given some images (each an |m x n x 3| array), show the images
    """
    n_rows = int(np.ceil(arr.shape[0] / n_cols))
    gridsize = n_rows * n_cols
    extras = np.arange(arr.shape[0], gridsize) 
    
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(16, 8))
    plt.tight_layout()

    for idx, imgarr in enumerate(arr):
        row, col = figpos(n_cols, idx)
        axs[row, col].imshow(imgarr)
    
```

```python
show(get_image(12), n_cols=4)
```
