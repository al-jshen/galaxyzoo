```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import torch
from torch import nn, optim
from torchvision import models, transforms, utils
from PIL import Image
from sklearn.model_selection import train_test_split
from tqdm import tqdm
```

```python
IMGPATH = 'images_training_rev1'
imglist = [f'{IMGPATH}/{i}' for i in os.listdir(IMGPATH)]
sols = np.loadtxt('training_solutions_rev1.csv', delimiter=',', skiprows=1)
```

```python
def rand_image(n=6):
    """ Get n random image filenames from training set
    """
    return np.random.choice(imglist, size=n, replace=False)
```

```python
def convert(arr):
    """ Given an array of filenames, return an array of corresponding numpy arrays, each |m x n x 3|\
    """
    return np.array([np.array(Image.open(f)) for f in arr])
```

```python
def figpos(n_cols, i):
    """ Given a number, return its corresponding coordinate in a 2D grid
    """
    row = i // n_cols
    col = i % n_cols
    
    return row, col

def show(arr, n_cols=3):
    """ Given some images (each an |m x n x 3| array), show the images
    """
    n_rows = int(np.ceil(arr.shape[0] / n_cols))
    gridsize = n_rows * n_cols
    extras = np.arange(arr.shape[0], gridsize) 
    
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(12, 6))
    plt.tight_layout()

    for idx, imgarr in enumerate(arr):
        row, col = figpos(n_cols, idx)
        im = axs[row, col].imshow(imgarr)
    
    for i in extras:
        row, col = figpos(n_cols, i)
        axs[row, col].remove()
    
```

```python
show(convert(rand_image(12)), n_cols=4)
```

```python
"""
Proceeding naively. Instead of following the flowchart (proper order), going to 
feed in the probabilities as a flat 
initial testing with a subset (first 1000 images)

"""


ids = sols[:1000,0].astype(int).astype(str)
train_ids, test_ids = train_test_split(ids, test_size=0.2, shuffle=True)
```

```python
def preprocess(ids):
    """ Given an array of ids as strings (int strings e.g. '2', '512'), 
    grab the corresponding features (images as 3D arrays) and labels.
    Also split the features and labels into training and testing sets.
    """
    filenames = [f'{IMGPATH}/{i}.jpg' for i in tqdm(ids)]
    features = convert(filenames)
    
    labels = np.array([sols[sols[:,0] == j][0][1:] for j in test_ids.astype(float)])
    
    return features, labels
```

```python
train_features, train_labels = preprocess(train_ids)
test_features, test_labels = preprocess(test_ids)
```
