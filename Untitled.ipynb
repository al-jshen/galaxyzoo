```python
import os
import csv
import copy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import torch
from torch import nn, optim
from torchvision import models, transforms, utils
from PIL import Image
from sklearn.model_selection import train_test_split
from tqdm.notebook import tqdm, trange
```

```python
IMGPATH = 'images_training_rev1'
imglist = [f'{IMGPATH}/{i}' for i in os.listdir(IMGPATH)]
sols = np.loadtxt('training_solutions_rev1.csv', delimiter=',', skiprows=1)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
def rand_image(n=6):
    """ Get n random image filenames from training set
    """
    return np.random.choice(imglist, size=n, replace=False)
```

```python
def convert(arr, fmt='pil'):
    """ Given an array of filenames, return an array of either PIL images or numpy arrays (each |m x n x 3|)
    
    """
    if fmt == 'pil':
        # have to do this copy workaround to prevent "too many files open" error
        return [copy.deepcopy(Image.open(f)) for f in arr]
    elif fmt == 'np':
        return np.array([np.array(Image.open(f)) for f in arr])
```

```python
def figpos(n_cols, i):
    """ Given a number, return its corresponding coordinate in a 2D grid
    """
    row = i // n_cols
    col = i % n_cols
    
    return row, col

def show(arr, n_cols=3):
    """ Given some images (each an |m x n x 3| array), show the images
    """
    n_rows = int(np.ceil(arr.shape[0] / n_cols))
    gridsize = n_rows * n_cols
    extras = np.arange(arr.shape[0], gridsize) 
    
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(12, 6))
    plt.tight_layout()

    for idx, imgarr in enumerate(arr):
        row, col = figpos(n_cols, idx)
        im = axs[row, col].imshow(imgarr)
    
    for i in extras:
        row, col = figpos(n_cols, i)
        axs[row, col].remove()
    
```

```python
"""
Proceeding naively. Instead of following the flowchart (proper order), going to make the model answer all questions as once. 
Probably a really bad idea since some paths are unreachable given particular previous conditions. 

Initial testing with a really small subset because not sure if itll work + super weak computer cant handle big stuff
"""

ids = sols[:100,0].astype(int).astype(str)
train_ids, test_ids = train_test_split(ids, test_size=0.2, shuffle=True)
```

```python
def preprocess(ids):
    """ Given an array of ids as strings (int strings e.g. '2', '512'), 
    grab the corresponding features (images as 3D arrays) and labels.
    Also split the features and labels into training and testing sets.
    """
    filenames = [f'{IMGPATH}/{i}.jpg' for i in ids]
    features = convert(filenames)
    
    labels = np.array([sols[sols[:,0] == j][0][1:] for j in test_ids.astype(float)])
    
    return features, labels
```

```python
train_features, train_labels = preprocess(train_ids)
test_features, test_labels = preprocess(test_ids)
```

```python
# quick sanity check to make sure stuff is working properly

fig, ax = plt.subplots(1, 2)
ax[0].imshow(train_features[51])
ax[1].imshow(convert([f'{IMGPATH}/{train_ids[51]}.jpg'], fmt='np')[0])
```

```python
# returns a 4D tensor of size (10 x 3 x 224 x 224)

tfms = transforms.Compose([
    transforms.Resize(256), 
    transforms.TenCrop(224),
    transforms.Lambda(lambda crops: torch.stack([transforms.ToTensor()(crop) for crop in crops])),
    transforms.Lambda(lambda crops: torch.stack([transforms.Normalize(mean=nrml_mean, std=nrml_std)(crop) for crop in crops]))
])
```

```python
class ds(torch.utils.data.Dataset):
    def __init__(self, features, labels, transform=None):
        self.features = features
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return(len(self.labels))

    def __getitem__(self, index):
        feature = self.features[index]
        feature = self.transform(feature)
        label = self.labels[index]
        return(feature, label)
```

```python
train_ds = ds(train_features, train_labels, tfms)
test_ds = ds(test_features, test_labels, tfms)
```

```python
train_loader = torch.utils.data.DataLoader(train_ds, batch_size=2, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_ds, batch_size=2, shuffle=True)
```

```python
model = models.resnet50(pretrained=True)
```

```python
model.fc
```

```python
class galaxynet(nn.Module):
    def __init__(self):
        super().__init__()        
        self.fc1 = nn.Linear(2048, 8192)
#         self.fc2 = nn.Linear(8192, 16384) # b i g net (12346)
#         self.fc3 = nn.Linear(16384, 4096)
#         self.fc4 = nn.Linear(4096, 1024)
        self.fc5 = nn.Linear(8192, 1024) #alt path (156)
        self.fc6 = nn.Linear(1024, 37)
        
        
        self.drop = nn.Dropout(p=0.6)
        self.relu = nn.LeakyReLU()

    def forward(self, x):
        x = self.drop(self.relu(self.fc1(x)))
#         x = self.drop(self.relu(self.fc2(x)))
#         x = self.drop(self.relu(self.fc3(x)))
#         x = self.drop(self.relu(self.fc4(x)))
        x = self.drop(self.relu(self.fc5(x)))
        x = self.fc6(x)
        return x
```

```python
clf = galaxynet()
model.fc = clf
model.fc
```

```python
# checking to see that the model works
feature, target = next(iter(train_loader))
bs, ncrops, c, h, w = feature.size()
feature = feature.view(-1, c, h, w)
output = model(feature)
output = output.view(bs, ncrops, -1)
output.mean(1)
```

```python
optimizer = optim.Adam(model.parameters(), lr=1e-4)
loss_fn = nn.MSELoss()

train_losses = []
test_losses = []
```

```python
epochs = 5

for e in trange(epochs, desc='Total training'):
    model.train()
    train_loss = 0
    for feature, target in tqdm(train_loader, desc=f'Training for epoch {e+1}'):
        optimizer.zero_grad()
        bs, ncrops, c, h, w = feature.size()
        feature = feature.view(-1, c, h, w)
        output = model(feature)
        output = output.view(bs, ncrops, -1)
        output = output.mean(1)
        loss = loss_fn(output, target.float())
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    
    model.eval()
    test_loss = 0
    for feature, target in tqdm(test_loader, desc=f'Testing for epoch {e+1}'):
        bs, ncrops, c, h, w = feature.size()
        feature = feature.view(-1, c, h, w)
        output = model(feature)
        output = output.view(bs, ncrops, -1)
        output = output.mean(1)
        loss = loss_fn(output, target.float())
        test_loss += loss.item()
    
    print(f'Epoch {e+1} | Training loss: {train_loss} | Testing loss: {test_loss}')
    
    train_losses.append(train_loss)
    test_losses.append(test_loss)
    
    if len(test_losses) == 1:
        torch.save(model, f'galaxyzoo.pt')
    else:
        if test_loss <= test_losses[-1]:
            torch.save(model, f'galaxyzoo.pt')
```
